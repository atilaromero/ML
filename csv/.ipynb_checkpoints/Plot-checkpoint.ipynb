```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def readfile(p):
    x = []
    y = []
    with open(p) as f:
        f.readline()
        for line in f:
            t,step,val_acc = line.strip().split(',')
            x.append(float(t))
            y.append(float(val_acc))
    x = np.array(x)
    y = np.array(y)
    x = x - x[0]
    x = x/60
    return x,y
```

```python
data = {}
for e in exps:
    data[e] = readfile('run-' + e + '-tag-epoch_val_acc.csv')
```

```python
def plot(experiments, loc='lower right'):
    fig = plt.figure(figsize=(12,8))
    for i,j in zip(experiments, ['--k', '-bx', '-g+', '-r*', '-mD']):
        plt.plot(*data[i], j, label=i)
    plt.ylim((0.32, 0.92))
    plt.xlim((-0.2, 10.5))
    plt.legend(loc=loc, prop={'size': 16})
    plt.grid()
    plt.savefig('-'.join(experiments)+'.png')
    plt.show()
```

```python
def readacc(p):
    x = []
    y = []
    with open(p) as f:
        f.readline()
        for line in f:
            n, val_acc = line.strip().split(' ')
            x.append(float(n))
            y.append(float(val_acc))
    x = np.array(x)
    y = np.array(y)
    return x,y
```

```python
x, y = readacc('acc-26-0.tsv')
```

```python
plt.plot(x,y, 'o')
plt.grid()
```

```python
x, y = readacc('acc-purif2-30.csv')
plt.plot(x,y, 'o')
plt.grid()
```

```python
from matplotlib.ticker import MultipleLocator
def plot_acc(filenames, loc='upper right'):
    fig = plt.figure(figsize=(12,8))
#     fig.yaxis.set_minor_locator()
    for filename,mark in zip(filenames, ['bx', 'g+', 'r*', '.mD']):
        x, y = readacc(filename)
        plt.plot(x, y, mark, label=filename)
    ax1 = plt.gca()
    plt.ylabel('accuracy', fontsize=16)
    plt.xlabel('number of classes', fontsize=16)
    plt.xlim((-0, 40))
    plt.ylim((-0, 1))
    ax1.xaxis.set_major_locator(MultipleLocator(5))
    ax1.xaxis.set_minor_locator(MultipleLocator(1))
    ax1.yaxis.set_major_locator(MultipleLocator(0.1))
    ax1.yaxis.set_minor_locator(MultipleLocator(0.05))
    plt.grid(linestyle='-', color='k', which='major')
    plt.grid(linestyle='--', which='minor')
    for tick in ax1.xaxis.get_major_ticks():
        tick.label.set_fontsize(16) 
    for tick in ax1.yaxis.get_major_ticks():
        tick.label.set_fontsize(16) 
    plt.legend(loc=loc, prop={'size': 16})
    plt.savefig('acc.png')
    plt.show()
```

```python
plot_acc(['acc-purif2-30.csv', 'acc-26-0.tsv'])
```
