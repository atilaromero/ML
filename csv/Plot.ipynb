```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def readfile(p):
    x = []
    y = []
    with open(p) as f:
        f.readline()
        for line in f:
            t,step,val_acc = line.strip().split(',')
            x.append(float(t))
            y.append(float(val_acc))
    x = np.array(x)
    y = np.array(y)
    x = x - x[0]
    x = x/60
    return x,y
```

```python
x,y = readfile('./run-2-all_tboard_CL-tag-epoch_val_acc.csv')
```

```python
plt.plot(x,y,'o')
plt.grid()
```

```python
exps = ['CCLL', 'CCL', 'CCM', 'CD', 'CLD', 'CLL', 'CL', 'CMCMLL', 'CMCML', 'CML', 'CM', 'D', 'LD']
```

```python
data = {}
for e in exps:
    data[e] = readfile('run-2-all_tboard_' + e + '-tag-epoch_val_acc.csv')
```

```python
def plot(experiments):
    fig = plt.figure(figsize=(12,8))
    for i,j in zip(experiments, ['--k', '-bx', '-g+', '-r*', '-mD']):
        plt.plot(*data[i], j, label=i)
    plt.ylim((0.32, 0.92))
    plt.xlim((-0.2, 10.5))
    plt.legend(loc='lower right')
    plt.grid()
    plt.savefig('-'.join(experiments)+'.png')
    plt.show()
```

```python
plot(['CL', 'D', 'LD'])
```
